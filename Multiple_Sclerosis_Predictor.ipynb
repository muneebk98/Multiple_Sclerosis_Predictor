### Dataset Overview
The dataset we've chosen is focused on predicting the conversion from Clinically Isolated Syndrome (CIS) to Multiple Sclerosis (MS). It comes from Kaggle and is aimed at identifying factors that might influence the likelihood of CIS patients progressing to MS.
https://www.kaggle.com/datasets/desalegngeb/conversion-predictors-of-cis-to-multiple-sclerosis/data

### Features in the Dataset
Here's a brief description of each feature in the dataset, including the type of data (numerical, categorical, or text):

1. **ID**: A unique numerical identifier for each patient. *(Numerical)*

2. **Age**: The age of the patient in years. *(Numerical)*

3. **Schooling**: The total number of years the patient spent in educational institutions. *(Numerical)*

4. **Gender**: Patient's gender, where '1' represents male and '2' represents female. *(Categorical)*

5. **Breastfeeding**: Indicates breastfeeding status, with '1' for yes, '2' for no, and '3' for unknown. *(Categorical)*

6. **Varicella (Chickenpox)**: Reflects history of chickenpox, with '1' for positive, '2' for negative, and '3' for unknown. *(Categorical)*

7. **Initial_Symptoms**: Categorizes initial symptoms into 15 types ranging from individual symptoms to combinations of visual, sensory, motor, and other symptoms. *(Categorical)*

8. **Mono_or_Polysymptomatic**: Specifies if the clinical presentation is monosymptomatic ('1'), polysymptomatic ('2'), or unknown ('3'). *(Categorical)*

9. **Oligoclonal_Bands**: Used to detect bands of immunoglobulins in cerebrospinal fluid, with '0' for negative, '1' for positive, and '2' for unknown. *(Categorical)*

10. **LLSSEP (Lower Limb SSEP)**: Measures somatosensory evoked potentials from the lower limbs, with '0' for negative and '1' for positive. *(Categorical)*

11. **ULSSEP (Upper Limb SSEP)**: Measures somatosensory evoked potentials from the upper limbs, with '0' for negative and '1' for positive. *(Categorical)*

12. **VEP (Visual Evoked Potential)**: Records the electrical activity in response to visual stimuli, with '0' for negative and '1' for positive. *(Categorical)*

13. **BAEP (Brainstem Auditory Evoked Potential)**: Measures auditory nerve responses, with '0' for negative and '1' for positive. *(Categorical)*

14. **Periventricular_MRI**: MRI findings around the brain's ventricles, indicating '0' for negative findings and '1' for positive findings. *(Categorical)*

15. **Cortical_MRI**: MRI findings in the cortical areas of the brain, with '0' for negative and '1' for positive. *(Categorical)*

16. **Infratentorial_MRI**: MRI findings in the infratentorial regions of the brain, with '0' for negative and '1' for positive. *(Categorical)*

17. **Spinal_Cord_MRI**: MRI findings related to the spinal cord, with '0' for negative and '1' for positive. *(Categorical)*

18. **Initial_EDSS**: The initial score on the Expanded Disability Status Scale, quantifying the degree of disability in the patient. *(Numerical)*

19. **Final_EDSS**: The final score on the EDSS after a period of observation or treatment, representing changes in disability status. *(Numerical)*

20. **Group**: The classification of the patient's condition, indicating whether they developed CDMS ('1') or not ('2'). *(Categorical)*



```python
import pandas as pd
df = pd.read_csv('data.csv')
if 'Unnamed: 0' in df.columns:                                                             #Dropping the first column "id" which is unnamed in this case
    df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df.describe()
```

```python
categorical_vars = ['Breastfeeding', 'Varicella', 'Mono_or_Polysymptomatic', 'Gender']      # Applying one-hot encoding to categorical variables that are non-ordinal
df = pd.get_dummies(df, columns=categorical_vars, drop_first=True)
print(df.shape)

# Return any columns that might still need addressing based on null values or other criteria
df.isnull().sum(), df.info()
```

```python
df.columns
```

```python
import numpy as np
df.drop(['Initial_EDSS', 'Final_EDSS'], axis=1, inplace=True)                         # Dropping due to high correlation and a large prectage of nulls.


df['Schooling'].fillna(df['Schooling'].median(), inplace=True)                        # Imputing missing values with median for numerical columns
df['Initial_Symptom'].fillna(df['Initial_Symptom'].median(), inplace=True)

Q1 = df.select_dtypes(include=['float64', 'int64']).quantile(0.25)                    # Detecting outliers using the IQR range and capping
Q3 = df.select_dtypes(include=['float64', 'int64']).quantile(0.75)
IQR = Q3 - Q1

exclude_columns = ['BAEP', 'VEP', 'LLSSEP', 'ULSSEP']
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    if column not in exclude_columns:
        lower_bound = Q1[column] - 1.5 * IQR[column]
        upper_bound = Q3[column] + 1.5 * IQR[column]
        df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
        df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])

print(df.head())

```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')

# Ensure you include the correct features after one-hot encoding. Example features listed here:
y = df['group']
x = df[['Age', 'Schooling', 'Initial_Symptom', 'Oligoclonal_Bands', 'LLSSEP',
       'ULSSEP', 'VEP', 'BAEP', 'Periventricular_MRI', 'Cortical_MRI',
       'Infratentorial_MRI', 'Spinal_Cord_MRI', 'Breastfeeding_2',
       'Breastfeeding_3', 'Varicella_2', 'Varicella_3',
       'Mono_or_Polysymptomatic_2', 'Mono_or_Polysymptomatic_3', 'Gender_2']]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", classification_rep)
```
